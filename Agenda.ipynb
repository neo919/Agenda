```python
agenda = {}  # Agenda funcional - Eng. Soft (Atividade Ativa)

def inserir (): # Bloco de códigos para inserir um contato
    nome = input("Insira o nome: ")
    telefone = input("Agora o telefone: ")
    email = input("O e-mail: ")
    twitter = input("O twitter: ")
    instagram = input("Por fim o Instagram: ")
    
    contato = {
        "Nome": nome,
        "Telefone": telefone,
        "E-mail": email,
        "Twitter": twitter,
        "Instagram": instagram
    }
    
    agenda[nome] = contato
    print("Salvo")

def dois_ou_mais(): #Função para inserir dois ou mais contatos
    quantos = int(input("Quantos contatos? "))
    for _ in range(quantos):
        inserir()

def consultar(): #Bloco de codigos para consultar um contato salvo
    nome = input("Insira o nome para consulta: ")
    if nome in agenda:
        contato = agenda[nome]
        for chave, valor in contato.items():
            print(f"{chave}: {valor}")

def deletar(): #Bloco de códigos para deletar um contato salvo
    nome = input("Contato a ser deletado: ")
    if nome in agenda:
        del agenda[nome]
        print("Deletado")

def modificar(): #Bloco de códigos para modificar um contato salvo
    nome = input("Contato a ser modificado: ")
    if nome in agenda:
        contato = agenda[nome]

        for chave, valor in contato.items():
            print(f"{chave}: {valor}")
        
        telefone = input("Novo telefone: ")
        email = input("Novo e-mail: ")
        twitter = input("Novo twitter: ")
        instagram = input("Novo instagram: ")
        
        contato["Telefone"] = telefone
        contato["E-mail"] = email
        contato["Twitter"] = twitter
        contato["Instagram"] = instagram
        
        print("Modificado")

def relatorio(): #Função que gera um relatório de todos contatos salvos
    print("Relatório:")
    for nome, contato in agenda.items():
        print(f"Nro: {len(agenda)}")
        print(f"Nome: {contato['Nome']}")
        print(f"E-mail: {contato['E-mail']}")
        print(f"Twitter: {contato['Twitter']}")
        print(f"Instagram: {contato['Instagram']}")
        print()

def salvar(): #Função que salva como arquivo os contatos
    arquivo = input("Insira um nome para o arquivo: ")

    try:
        with open(arquivo, "w") as f:
            f.write("Nome,Telefone,E-mail,Twitter,Instagram\n")
            for contato in agenda.values():
                nome = contato["Nome"]
                telefone = contato["Telefone"]
                email = contato["E-mail"]
                twitter = contato["Twitter"]
                instagram = contato["Instagram"]
                linha = f"{nome},{telefone},{email},{twitter},{instagram}\n"
                f.write(linha)
        print("Salvo")
    except IOError:
        print("Erro")

def exibir_menu(): #Menu com todas as funções
    print("Menu:")
    print("1. Inserir")
    print("2. Inserir 2 ou mais")
    print("3. Consultar")
    print("4. Deletar")
    print("5. Modificar")
    print("6. Relatório")
    print("7. Salvar contatos")

exibir_menu()

while True:

    opcao = input("Digite a opção desejada: ")
    if opcao == "1":
        inserir()
    elif opcao == "2":
          dois_ou_mais()
    elif opcao == "3":
          consultar()
    elif opcao == "4":
          deletar()
    elif opcao == "5":
          modificar()
    elif opcao == "6":
          relatorio()
    elif opcao == "7":
          salvar()
          break

```
